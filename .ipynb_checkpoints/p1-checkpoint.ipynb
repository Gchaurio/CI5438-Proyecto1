```python
import pandas as pd
import numpy as np

df = pd.read_csv("car_data_preprocessed.csv")

cs = ['Make', 'Year', 'Kilometer', 'Fuel Type', 'Transmission', 'Owner', 'Seating Capacity', 'Fuel Tank Capacity']
cols = []
for i in list(df.columns)[1:]:
    for j in cs:
        if j in i:
            cols.append(i)
```

```python
Y = df.loc[:,"Price"]
X = df.loc[:,cols]
X.insert(0, "termino_indep", [1.0]*len(X))

W = [1.0 for i in X]

for i in range(10000):
    hw = np.dot(W, X.T)
    err = Y - hw

    W = W + (0.0001 * np.dot(X.T, err) * (2/len(X)))
```
